```python
import os
import docx
```

```python
def readtxt(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return fullText

text =  (readtxt("D:/Profiles/sopra/Downloads/Resume Template.docx"))
```

```python
text = [t.strip() for t in text if t.strip()!= '']   #removing excess whitespaces
```

```python
text_ = [t.split("  ") for t in text]   #split if there are any excess whitespaces
```

```python
text
```

```python
text_2  =[sentence for LIST in text_ for sentence in LIST]
```

```python
text_2 = [t.strip() for t in text_2 if t.strip()!= '']  #remove ""
```

```python
text_2 = [t.split('\t') for t in text_2]   #remove "\t"
```

```python
text_2  =[sentence for LIST in text_2 for sentence in LIST]       
```

```python
text_2 = [t.strip() for t in text_2 if t.strip()!= '']             #removed excess spaces
```

```python
text_2
```

```python
import nltk
```

```python
tokenized  = [nltk.word_tokenize(sentence) for sentence in text_2]   #tokenized words
```

```python
index  = []
for i,j in enumerate(tokenized):
        index.extend([i+1]*len(j))      #labelling words with sentence number
```

```python
index
```

```python
final = []
for sentence in tokenized:
    final.extend(sentence)
```

```python
len(final)
```

```python
len(index)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df['Sentence_No'] = index
df['Word'] = final
```

```python
df.head(3)
```

```python
tags = [nltk.pos_tag(sentence) for sentence in tokenized]    #part of speech tagging
```

```python
POS = []
for sentence in tags:
    for _,tag in sentence:
        POS.append(tag)
```

```python
len(POS)
```

```python
df['POS'] = POS
```
